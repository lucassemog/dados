```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregue a planilha e especifique a aba "Municipios"
planilha = pd.read_excel('Estado_PR.xlsx', sheet_name='Municipios')

# Converta os valores da coluna 'Mortalidade infantil - por mil nascidos vivos [2020]' para numéricos e substitua erros por NaN
planilha['Mortalidade infantil - por mil nascidos vivos [2020]'] = pd.to_numeric(planilha['Mortalidade infantil - por mil nascidos vivos [2020]'], errors='coerce')

# Ordene a planilha pela coluna de mortalidade infantil em ordem decrescente
mortalidade_infantil_2020 = 'Mortalidade infantil - por mil nascidos vivos [2020]'
planilha_ordenada = planilha.sort_values(by=mortalidade_infantil_2020, ascending=False)

# Selecione as 10 cidades com maior taxa de mortalidade infantil
top_10_mortalidade = planilha_ordenada.head(10)

# Crie o gráfico de barras
plt.figure(figsize=(12, 6))
plt.barh(top_10_mortalidade['Municipio'], top_10_mortalidade[mortalidade_infantil_2020], color='skyblue')
plt.xlabel('Mortalidade Infantil (por mil nascidos vivos) - 2020')
plt.ylabel('Município')
plt.title('Top 10 Municípios com Maior Mortalidade Infantil em 2020')
plt.gca().invert_yaxis()  # Inverte o eixo y para classificar do maior para o menor

# Exiba o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregue o arquivo Excel
caminho_arquivo = 'Estado_PR.xlsx'

# Leia o arquivo Excel e especifique a página "Idade_sexo_2020"
df = pd.read_excel(caminho_arquivo, sheet_name='Idade_sexo_2020')

# Defina as faixas etárias como índices (se já não estiverem)
df.set_index('Faixa Etária', inplace=True)

# Selecione as colunas relevantes: "Total", "Homens" e "Mulheres"
colunas_selecionadas = df[['Total', 'Homens', 'Mulheres']]

# Crie o gráfico de barras
ax = colunas_selecionadas.plot(kind='bar', figsize=(12, 6))
plt.xlabel('Faixa Etária')
plt.ylabel('População')
plt.title('Distribuição da População por Faixa Etária (2020)')
plt.legend(title='Sexo')
plt.xticks(rotation=45)  # Rotação dos rótulos no eixo x para melhor legibilidade

# Exiba o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregue o arquivo Excel e selecione a página "Veículos"
caminho_arquivo = 'Estado_PR.xlsx'
xls = pd.ExcelFile(caminho_arquivo)

# Verifique as colunas disponíveis na página "Veículos"
print("Colunas disponíveis na página 'Veículos':")
print(xls.parse('Veículos').columns)

# Selecione as colunas relevantes para o cálculo da soma total
colunas_veiculos = ['Automóvel ', 'Caminhão ', 'ônibus ', 'Motocicleta ']

# Calcule a soma total em cada categoria
df = xls.parse('Veículos')
totais = df[colunas_veiculos].sum()

# Crie rótulos para as categorias
categorias = colunas_veiculos

# Crie o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(totais, labels=categorias, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Veículos por Categoria')
plt.axis('equal')  # Para garantir que o gráfico de pizza seja uma forma circular

# Exiba o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregue a planilha e especifique a aba "Municipios"
planilha = pd.read_excel('Estado_PR.xlsx', sheet_name='Municipios')

# Ordene a planilha pela coluna de população em ordem decrescente
planilha_ordenada = planilha.sort_values(by='População residente -[2022]', ascending=False)

# Selecione as 10 primeiras linhas (as 10 maiores cidades)
top_10_maiores_cidades = planilha_ordenada.head(10)

# Crie um gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(top_10_maiores_cidades['População residente -[2022]'], labels=top_10_maiores_cidades['Municipio'], autopct='%1.1f%%', startangle=140)
plt.title('População das 10 Maiores Cidades do Paraná em 2022')

# Crie uma legenda
legenda = []
for index, row in top_10_maiores_cidades.iterrows():
    legenda.append(f"{row['Municipio']} - {row['População residente -[2022]']} habitantes")

plt.legend(legenda, loc='best', bbox_to_anchor=(1, 0.5))

# Exiba o gráfico
plt.axis('equal')  # Mantém o gráfico de pizza em formato de círculo
plt.show()
```
